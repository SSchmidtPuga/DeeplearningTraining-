```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create features, independent variables
X = np.array([-7.0,-4.0,-1.0,2.0,5.0,8.0,11.0,14.0, 17.0,20.0,23.0])

#create labels, dependent variable 
y = np.array([3.0,6.0,9.0,12.0,15.0,18.0,21.0,24.0, 27.0,30.0,33.0])

#Visualize it
plt.scatter(X,y)
```

```python
#Demo tensor  for housing price prediction problem
X = tf.cast(tf.constant(X),dtype= tf.float32)
y = tf.cast(tf.constant(y),dtype= tf.float32)
X,y


```

# Steps in modelling with Tensorflow 
1) Creatien a model: Define the input and output layers, as we as the hidden layers of a deep learning model.

2) compiling a model -  define the loss function (In other words, the function which tells our model how wrong it is) and optimize (tells our model how to improve the patterns its learning) with evaluation metrics ( What we use to interpret the performance of our model.

3) Fitting a model - letting the model try to find patterns between X & y (Features and labels



```python
# Data into tensors 

X = tf.constant(X)
y = tf.constant(y)

# 1. Create a model using the sequential API
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(1)])

# 2. Compile

#mae is mean absolute error -> Media del error en predicciones vesus la realidad -> En promedio que tan mal esta nuestra prediccion
#SGD -> Stochastic grandient descent
model.compile(loss = tf.keras.losses.mae, 
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"]) 


#3. Fit the model
#Epochs es el numero de intentos que le damos 

model.fit(tf.expand_dims(X, axis=-1), y, epochs = 200)


```

```python
#try and make a pediction using our model
numero = model.predict([17.0])
numero
```

Improving the model
# Steps in Improve a modelmodelling with Tensorflow 

1)Re-Creating the model 2.0 : Add more layers, increase the number of hideen units(all called neurons) within each of the hideen layers , change the activation function.

2)Re-Compiling the model -  Change the omptimization function or perhaps the learning rate of the function

3)Re-Fitting the model - Fit the model for more epochs( More trys) or give more data to the model

```python
# Data into tensors 

X = tf.constant(X)
y = tf.constant(y)

# 1. Create a model using the sequential API
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(100, activation = "relu"),
                             tf.keras.layers.Dense(100, activation = "relu"),
                             tf.keras.layers.Dense(100, activation = "relu"),
                             tf.keras.layers.Dense(1)])

# 2. Compile

#mae is mean absolute error -> Media del error en predicciones vesus la realidad -> En promedio que tan mal esta nuestra prediccion
#SGD -> Stochastic grandient descent
model.compile(loss = tf.keras.losses.mae, 
              optimizer = tf.keras.optimizers.Adam(lr = 0.01),
              metrics = ["mae"]) 


#3. Fit the model
#Epochs es el numero de intentos que le damos 
model.fit(tf.expand_dims(X, axis=-1), y, epochs = 100)
```

```python
X = tf.range(-100,100,4)
y = X+10
```

```python
# Data into tensors 

X = tf.constant(X)
y = tf.constant(y)

# 1. Create a model using the sequential API
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(100, activation = "relu"),
                             tf.keras.layers.Dense(100, activation = "relu"),
                             tf.keras.layers.Dense(100, activation = "relu"),
                             tf.keras.layers.Dense(1)])

# 2. Compile

#mae is mean absolute error -> Media del error en predicciones vesus la realidad -> En promedio que tan mal esta nuestra prediccion
#SGD -> Stochastic grandient descent
model.compile(loss = tf.keras.losses.mae, 
              optimizer = tf.keras.optimizers.Adam(lr = 0.01),
              metrics = ["mae"]) 


#3. Fit the model
#Epochs es el numero de intentos que le damos 

model.fit(tf.expand_dims(X, axis=-1), y, epochs = 1000,verbose = 1)
```

```python
#Split the data into train and test sets
X_train = X[:40] # First 40 -> 80% of data
y_train = y[:40]

X_test = X[40:] # Last 10 -> 20% of data
y_test = y[40:] 
```

```python
plt.figure(figsize=(10,7))

plt.scatter(X_train, y_train,c="b", label ="Training data")
plt.scatter(X_test, y_test,c="g", label ="test data")
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(100, input_shape = [1], name= "input_layer"),
                              tf.keras.layers.Dense(100, input_shape = [1], name= "Output_layer")])

model.compile(loss= tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics=["mae"])

model.fit(tf.expand_dims(X, axis=-1), y, epochs = 100, verbose=0)
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model, show_shapes=True)
```

```python
y_preds = model.predict(X_test)
y_preds 
```

```python
def plot_predictions(train_data = X_train,
                     train_labels=y_train,
                     test_data= X_test,
                     test_labels=y_test,
                     predictions =y_preds):
  plt.figure(figsize=(10,7))
  plt.scatter(train_data, train_labels,c="b", label ="Training data")
  plt.scatter(test_data, test_labels,c="g", label ="test data")
  plt.scatter(test_data, predictions, c="r", label="predictions")

plot_predictions()

```

```python
model.evaluate(X_test,y_test)
```

```python
tf.metrics.mean_absolute_error(y_test,tf.squeeze(tf.constant(y_preds)))
tf.metrics.mean_absolute_error(y_test,tf.squeeze(tf.constant(y_preds)))
```

```python
def mae_error(y_test = y_test,y_preds= y_preds ):
  return tf.metrics.mean_absolute_error(y_test,tf.squeeze(tf.constant(y_preds)))

mae_error() 
```
