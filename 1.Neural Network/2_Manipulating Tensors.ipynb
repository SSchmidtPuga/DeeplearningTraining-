```python
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
#find the positional maximum
tf.argmax(F)
```

```python
#index on our largest value
F[tf.argmax(F)]
```

```python
#Max value of F
tf.reduce_max(F)
```

```python
tf.argmin(F)
F[tf.argmin(F)]
tf.reduce_min(F)
```
